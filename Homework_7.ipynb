##Exe 7.1

```python
class Vector:

    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    def __repr__(self):
        return f"Vector({self.x}, {self.y}, {self.z})"

    def __eq__(self, other):
        if self.x == other.x and self.y == other.y and self.z == other.z:
            return True
        else:
            return False

    def __ne__(self, other):
        return not self == other

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y, self.z + other.z)

    def __sub__(self, other):
        return Vector(self.x - other.x, self.y - other.y, self.z - other.z)

    def __mul__(self, other):
        return (self.x * other.x + self.y * other.y + self.z * other.z)

    def cross(self, other):
        return Vector(self.y * other.z - self.z * other.y,
                      self.z * other.x - self.x * other.z,
                      self.x * other.y - self.y * other.x)

    def length(self):
        return math.sqrt(self.x ** 2 + self.y ** 2 + self.z ** 2)

    def __hash__(self):
        return hash((self.x, self.y, self.z))

    def find_axis(self, other):
      v3 = self.cross(other)
      if v3.length() == 0:
        raise ValueError("Vectors are parallel")

      length = v3.length()
      return Vector(v3.x/length, v3.y/length, v3.z/length)
```

```python
v1 = Vector(1, 0, 0)
v2 = Vector(0, 1, 0)

axis = v1.find_axis(v2)

assert axis == Vector(0, 0, 1)
assert axis.length() == 1.0

print("Test passed")
```

## Exe 7.2

```python
def test_iter(iter_name):
  for _ in range(10):
    print(next(iter_name))
```

(a) returning 0, 1, 0, 1, 0, 1, ...,

```python
a = itertools.cycle(range(0, 2))
```

```python
test_iter(a)
```

(b) returning randomly 0 or 1 on demand,


```python
b = iter(lambda: random.randint(0, 1), None)
```

```python
test_iter(b)
```

(c) returning 0, 1, 0, -1, 0, 1, 0, -1, ...

```python
c = itertools.cycle([0, 1, 0, -1])
```

```python
test_iter(c)
```
