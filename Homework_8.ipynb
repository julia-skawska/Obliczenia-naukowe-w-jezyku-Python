##Exe 8.1

```python
def test_array(array):
  print(array)
  print(array.shape)
  print(array.ndim)
  print(array.dtype)
```

```python
a = np.linspace(0, 1, 11, float)

test_array(a)
```

```python
b = np.zeros((5, 6), dtype=int)

test_array(b)
```

```python
x = complex(0, 1)
c = x ** np.arange(9)

test_array(c)
```

## Exe 8.2

```python
v1 = np.array([9, 12, 2001, 11, 11, 2007])
print(f"v1:\n {v1}")

v2 = v1[1::2]
print(f"v2:\n {v2}")

v3 = v1[::-1]
print(f"v3:\n {v3}")
```
