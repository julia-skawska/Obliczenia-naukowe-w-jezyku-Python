```python
words = S.split()

longest_word = [""]
for word in words: 
    if len(word) > len(longest_word[0]):
        longest_word = []
        longest_word.append(word)
    elif len(word) == len(longest_word[0]):
        longest_word.append(word)

print(f"Number of words in text: {len(words)}")
print(f"Longest word/s: {longest_word}")
```

## Exe 2.6

```python
t = (2, 4)
```

```python
print(t[2])
```

Otrzymujemy błąd indeksu: "tuple index out of range", znaczy to że element krotki który chcemy wypisać jest poza zasięgiem,
Wynika to z tego że w pythonie indeksy liczone są od 0, zatem 2 odpowiada indeksowi 0 a 4 odpowiada indeksowi 1,
więc w krotce t indeks 2 jeszcze nie istnieje

```python
t.append(6)
```

nawiasty:
[] - oznaczają listę
() - oznaczją krotkę

krotka którą jest tutaj "t" pomimo zewnętrznego przypominania listy, jest innym typem danych i bazuje na innych zasadach.
Atrybut append który pozwala dołożyć element do listy, nie działa w krotce która w przeciwieństwie do listy jest niemutowalna.

Aby dołożyć elementy do krotki należy zmienić jej typ na listę, lub dodać w formie drugiej krotki

```python
a, b = t ; print(a, b)
```

Do nowych elementów a oraz b wpisywane są elementy z krotki:
a = t[0] a b = t[1], czyli a = 2 i b = 4
te elementy następnie są wypisywane jako a b, czyli 2 4

```python
roman_to_arabic_v1 = {
    "I": 1,
    "IV": 4,
    "V": 5,
    "IX": 9,
    "X": 10,
    "XL": 40,
    "L": 50,
    "XC": 90,
    "C": 100,
    "CD": 400,
    "D": 500,
    "CM": 900,
    "M": 1000
}

print(f"Conversion from roman to arabic: IX is {roman_to_arabic_v1["IX"]}")
```

```python
arabic = ["I", "IV", "V", "IX", "X", "XL", "L", "XC", "C", "CD", "D", "CM", "M"]
roman   = [1, 4, 5, 9, 10, 40, 50, 90, 100, 400, 500, 900, 1000]

roman_to_arabic_v2 = dict(zip(arabic, roman))
print(f"Conversion from roman to arabic: CM is {roman_to_arabic_v2["CM"]}")
```
