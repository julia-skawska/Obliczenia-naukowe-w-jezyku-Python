## Exe 5.1

```python
def find_pdf_size(top):
    total_size = 0

    for root, dirs, files in os.walk(top):
        for file in files:
            if file.endswith(".pdf"):
                path = os.path.join(root, file)
                total_size += os.path.getsize(path)

    print(total_size)
    return None
```

## Exe 5.2

```python
def print_working_days(date1, date2):
    date1 = datetime.strptime(date1, "%Y-%m-%d")
    date2 = datetime.strptime(date2, "%Y-%m-%d")

    workdays = 0
    current = date1

    while current < date2:
        if current.weekday() < 5:
            workdays += 1
        current += timedelta(days=1)

    return workdays
```

```python
print(print_working_days("2001-12-09", "2007-11-11"))
```

```python
def random_walk(start):
  position = start
  while True:
      step = random.choice([-1, 1])
      position += step
      yield position
```

```python
print(random.randrange(-1, 1, 2))
```

```python
def test_final(steps):
    walk = random_walk(0)
    final_position = None
    for _ in range(steps):
        final_position = next(walk)
    return final_position
```

```python
random.seed(42)
print(test_final(100))
```
