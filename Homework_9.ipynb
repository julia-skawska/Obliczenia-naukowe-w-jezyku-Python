## Importy


```python
import numpy as np
import matplotlib.pyplot as plt
```

##Exe 9.1

```python
x = np.arange(0, 10, 0.1)
sin = np.sin(x)
cos = np.cos(x)
exp = np.exp(-x)

plt.plot(x, sin, color='red', label="sin(x)")
plt.plot(x, cos, color='green', linestyle="dashed", label="cos(x)")
plt.plot(x, exp, color='blue', linestyle="dotted", label="exp(-x)")
plt.legend(loc="upper right")
plt.show()
```

## Exe 9.2

```python
x = np.random.rand(100)
y = np.random.rand(100)

plt.figure(figsize=(5, 5))
dist = np.sqrt(x**2 + y**2)
col = ["green" if d < 1 else "red" for d in dist]

size = 10*(np.abs(x) + np.abs(y))
plt.scatter(x, y, c=col, s=size)
plt.grid(True)
plt.show()
```

## Exe 9.3


```python
x = np.linspace(-5, 5, 101)
y = np.linspace(-5, 5, 101)

X, Y = np.meshgrid(x, y)
Z = np.sqrt((X + 1)**2 + Y**2) + np.sqrt((X - 1)**2 + Y**2)

plt.imshow(Z, extent=[-5, 5, -5, 5], cmap="magma_r")
plt.colorbar()
plt.show()
```
