# Xenium Visualiser

Python GUI for the visualisation of Xenium spatial transcriptomics data stored in `.h5ad` (AnnData) format. The tool is designed to provide an easy workflow for researchers who want to explore spatial gene expression across different cell populations without requiring coding experience.

Users can load a Xenium (or other spatial transcriptomics) dataset, select a metadata column for grouping cells, specify genes of interest, and generate spatial expression plots.

The program combines **Tkinter** for the graphical interface, **Scanpy** for handling annotated biological datasets, and **Matplotlib** for visualisation.

## Features

* Import Xenium datasets stored as `.h5ad` files
* Automatically detect available metadata columns in `adata.obs`
* Select grouping variables (e.g. condition, cell type, sample)
* Visualise spatial distribution of one or multiple genes simultaneously (recommended ≤ 5 genes)
* Support for human and non-human gene naming conventions
* Automatic saving of generated plots

## Workflow

1. Load a Xenium `.h5ad` dataset (input path)
2. Select output directory for plots
3. Choose a metadata column for grouping cells
4. Enter one or more genes of interest
5. Generate spatial expression plots for each group
6. Navigate through results interactively
7. Automatically save figures for future reference

## Output

For each selected group, the program generates a spatial scatter plot where:

* All cells in the selected group are shown in the background (gray)
* Cells expressing selected genes are highlighted in distinct colours (one per gene)
* A legend is included for gene identification
* Multiple genes can be visualised simultaneously (recommended ≤ 5 genes)
* Figures are saved as `.png` files using the naming convention:
  `{anndata_file_name}_{grouping_variable}_{gene1}_{gene2}_..._{geneX}.png`

## Requirements

The environment is managed using the provided `SciProject.yml` file. Core dependencies include:
Scanpy, NumPy, Matplotlib, and Tkinter.

## Installation

### 1. Download the Project

Ensure the following structure is present:

```
Project/
│
├── Xenium_Visualiser.py
├── SciProject.yml
└── Images_For_UI/
    ├── Xvis.ico
    └── Xvis.png
```

### 2. Create the Conda Environment

```bash
conda env create -f SciProject.yml
```

### 3. Activate the Environment

```bash
conda activate SciProject
```

## Running the Program

```bash
python Xenium_Visualiser.py
```

## Using the Application

* Load dataset and output path (validated automatically)
* Select grouping variable from dropdown menu
* Enter genes of interest (checked against dataset)
* Generate plots
* Navigate results using next/previous controls

## Project Aim

The aim of this project is to provide an accessible tool for visualising spatial gene expression patterns within Xenium (spatial transcriptomics) datasets. By combining a user-friendly graphical interface with automated figure generation, it enables a simplified process for analysis and comparison of gene expression across biologically relevant cell groups.

## Limitations

* Currently only supports Xenium outputs after QC, clustering, and annotation
* Does not automatically handle different gene naming conventions
* Does not support user-defined plot customisation
